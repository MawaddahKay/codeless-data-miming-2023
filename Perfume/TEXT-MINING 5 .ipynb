```python
import re
```

```python
def remove_emojis(text):
    if isinstance(text, str):
        emoji_pattern = re.compile("["
            u"\U0001F600-\U0001F64F"
            u"\U0001F300-\U0001F5FF"
            u"\U0001F680-\U0001F6FF"
            u"\U0001F700-\U0001F77F"
            u"\U0001F780-\U0001F7FF"
            u"\U0001F800-\U0001F8FF"
            u"\U0001F900-\U0001F9FF"
            u"\U0001FA00-\U0001FA6F"
            u"\U0001FA70-\U0001FAFF"
            u"\U0001F004-\U0001F0CF"
            u"\U0001F170-\U0001F251"
            "]+", flags=re.UNICODE)
        return emoji_pattern.sub(r'', text)
    else:
        return text

# Apply the remove_emojis function to the specified column, ignoring non-string values
df['removeemoji'] = df['รายละเอียดสินค้า'].apply(lambda x: remove_emojis(x) if isinstance(x, str) else x)
```

```python
def remove_special_symbols(text):
    pattern = re.compile(r'[!@#$^&*()_+{}\[\]:;<>,?~\\|' ' '  ' ""]')
    
    # Use the pattern to remove special symbols from the text
    cleaned_text = pattern.sub('', text)
    
    return cleaned_text

df['removespecial'] = df['removeemoji'].apply(lambda x: remove_special_symbols(x) if isinstance(x, str) else x)
```

```python
from collections import Counter
import pandas as pd
import math

# Flatten the list of tokens
flat_tokens = [token for tokens in df['remove'] for token in tokens]

# Count term frequency
tf_counter = Counter(flat_tokens)

# Calculate the total number of documents
total_documents = len(df)

# Calculate IDF for each term
idf_values = {}
for term, tf in tf_counter.items():
    idf = math.log10(total_documents / (1 + tf))  # Using base 10 logarithm for IDF
    idf_values[term] = idf

# Add IDF values to the tf_df DataFrame
tf_df = pd.DataFrame(tf_counter.items(), columns=['Word', 'TF'])
tf_df['IDF'] = tf_df['Word'].map(idf_values)

# Calculate TF-IDF and add it to the DataFrame
tf_df['TF-IDF'] = tf_df['TF'] * tf_df['IDF']

# Sort by TF-IDF in descending order
tf_df = tf_df.sort_values(by='TF-IDF', ascending=False)

# Display the DataFrame with TF, IDF, and TF-IDF
tf_df
```

# LDA MODEL

```python
# Import necessary libraries
from gensim import corpora, models

# Tokenized and stop words removed data from the 'stopword' column
texts = df['stopword']

# Create a dictionary
dictionary = corpora.Dictionary(texts)

# Create a corpus (a bag of words representation)
corpus = [dictionary.doc2bow(text) for text in texts]

# Train the LDA model
lda_model = models.LdaModel(corpus, num_topics=10, id2word=dictionary, passes=15)

# Print the topics
topics = lda_model.print_topics(num_words=5)
for topic in topics:
    print(topic)
```
